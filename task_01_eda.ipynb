### Fix Date & Time Mandatory

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df.head(2)
```

### Feature Engineering

```python
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day_name()
df['Hour'] = df['Time'].dt.hour
```

## Gross Income Analysis by Product Line

- **Food and beverages** have the highest gross income, while **Health and beauty** has the lowest.
- Other lines like **Sports and travel** and **Electronic accessories** also contribute significantly.
- Shows that **high revenue doesn’t always mean high quantity sold**.


```python
df.groupby('Product line')['Quantity'].sum()
```
